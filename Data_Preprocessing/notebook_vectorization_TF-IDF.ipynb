```python
import numpy as np
import pandas as pd
from IPython.display import display


from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tweet_df = pd.read_excel ('../Data/preprocessed/NG_ELECTION_TWEETS_CLEANED.xlsx')

tfidf = TfidfVectorizer (min_df=.0005, ngram_range =(1,3))
tfidf.fit (tweet_df['cleaned_tweet'])
tfidf_tweet = tfidf.transform (tweet_df['cleaned_tweet'])

data = pd.DataFrame.sparse.from_spmatrix (tfidf_tweet, columns=tfidf.get_feature_names_out ())
data = pd.concat ([tweet_df ['Sentiment'], data], axis=1)

display (data.head (3))
```

```python
data.to_csv ('../Data/preprocessed/NG_TWEETS_TFIDF.csv', index=False)
```
